## Clean Dataframe

```python
df["Rating"] = df["Rating"].fillna(0)
df.isnull().sum()
```

### We will change the columns name and delete some columns

```python
df.rename(columns={df.columns[0]: "Id"}, inplace=True)
df.rename(mapper=lambda x: x.replace(" ", ""), axis=1, inplace=True)
```

```python
df.drop(columns=["Summary", "Reviews"], inplace=True)
df.columns
```

### We will change the dtype 

```python
cols_to_numbers = ["TimesListed", "NumberofReviews", "Plays", "Playing", "Backlogs", "Wishlist"]

df_copy_cols = df[cols_to_numbers].copy()
df.update(df_copy_cols.applymap(format_number))
```

```python
# For each column, we change the dtype
for col in cols_to_numbers:
    df[col] = df[col].astype("int64")

df.info()
```

```python
# We apply the function to the columns for cleaning strings
df["Genres"] = df["Genres"].apply(clean_string)
df["Team"] = df["Team"].apply(clean_string)
df[["Genres", "Team"]].head(5)
```

```python
# Sort the values by Whishlist
df = df.sort_values(by="Rating", ascending=False)
df.head()
```

```python
# We drop the duplicates values by Title
df.drop_duplicates(subset="Title", keep="first", inplace=True)
```
